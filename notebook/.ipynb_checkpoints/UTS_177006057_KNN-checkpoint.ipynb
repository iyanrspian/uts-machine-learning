```python
# Import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import dataset
link = "https://raw.githubusercontent.com/iyanrspian/uts-machine-learning/main/lokasi-site-bts.csv"
data = pd.read_csv(link)
data.head()
```

```python
# Update dataset yang akan diperlukan saja
data_update = pd.read_csv(link, usecols=['tinggi', 'luas tanah', 'tipe menara'])
data_update.head()
```

```python
# Menentukan variabel independen dan dependen
X = data_update.iloc[:, :-1].values
y = data_update.iloc[:, -1].values
```

```python
# Mengganti label 'tipe menara' menjadi integer
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

```python
# Split data training dan test (70:30)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 5)
```

```python
print(X_train)
```

```python
print(y_test)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
# Training KKN-Model pada data training (Euclidean Method)
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 10, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```python
# Prediksi hasil baru
print("Prediksi 1 : ", classifier.predict(scaler.transform([[30,50]])))
print("Prediksi 2 : ", classifier.predict(scaler.transform([[25,4]])))
print("Prediksi 3 : ", classifier.predict(scaler.transform([[45,125]])))

# Ket : (Tipe Menara)
# 0 = 3 Kaki
# 1 = 4 Kaki
# 2 = Monopole
# 3 = Pole
```

```python
# Prediksi hasil test set
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
# Probabilitas hasil prediksi
classifier.predict_proba(X_test)
```

```python
# Menghitung confusion matrix dan accuray score
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy = accuracy_score(y_test, y_pred)*100
print("\nConfusion matrix\n", confusion_matrix(y_test, y_pred))
print("\nAccuracy score :", str(round(accuracy, 2)) + '%')
```

```python
# Menampilkan hasil klasifikasi
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
# Cross Validation Methods
from sklearn.neighbors import KNeighborsClassifier
test_scores = []
train_scores = []

for i in range(1,10,1):
    knn = KNeighborsClassifier(i)
    knn.fit(X_train,y_train)
    
    train_scores.append(knn.score(X_train,y_train))
    test_scores.append(knn.score(X_test,y_test))

# Menghitung max_train_score
max_train_score = max(train_scores)
train_scores_ind = [i for i, v in enumerate(train_scores) if v == max_train_score]
print("Max train score {}% dan k = {}".format(max_train_score*100,list(map(lambda x: x+1, train_scores_ind))))

# Menghitung max_test_score
max_test_score = max(test_scores)
test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
print("Max test score {}% dan k = {}".format(round((max_test_score*100),2),list(map(lambda x: x+1, test_scores_ind))))
```

```python
# Visualisasi skor train set dan test set
import seaborn as sns

plt.figure(figsize=(10,4))
plt.xlabel('K Value')
plt.ylabel('Score')
line = sns.lineplot(data=train_scores, marker='*', label='Train Score')
line = sns.lineplot(data=test_scores, marker='o', label='Test Score')
```

```python
# Visualisasi hasil train set
from matplotlib.colors import ListedColormap

clmap = ListedColormap(['#FF0000', 'green', '#0000FF', 'orange'])
Xset, yset = scaler.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = Xset[:, 0].min() - 10, stop = Xset[:, 0].max() + 10, step = 1),
                     np.arange(start = Xset[:, 1].min() - 80, stop = Xset[:, 1].max() + 80, step = 1))

plt.contourf(X1, X2, classifier.predict(scaler.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.35, cmap = clmap)
for i, k in enumerate(np.unique(yset)):
    plt.scatter(Xset[yset == k, 0], Xset[yset == k, 1], color = clmap(i), label = k)
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
plt.title('KNN (Test set)')
plt.xlabel('Tinggi')
plt.ylabel('Luas Tanah')
plt.legend()
plt.show()

# Ket : (Tipe Menara)
# 0 = 3 Kaki
# 1 = 4 Kaki
# 2 = Monopole
# 3 = Pole
```

```python
# Visualisasi hasil test set
from matplotlib.colors import ListedColormap

clmap = ListedColormap(['#FF0000', 'green', '#0000FF', 'orange'])
Xset, yset = scaler.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = Xset[:, 0].min() - 10, stop = Xset[:, 0].max() + 10, step = 1),
                     np.arange(start = Xset[:, 1].min() - 80, stop = Xset[:, 1].max() + 80, step = 1))

plt.contourf(X1, X2, classifier.predict(scaler.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.35, cmap = clmap)
for i, k in enumerate(np.unique(yset)):
    plt.scatter(Xset[yset == k, 0], Xset[yset == k, 1], color = clmap(i), label = k)
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
plt.title('KNN (Test set)')
plt.xlabel('Tinggi')
plt.ylabel('Luas Tanah')
plt.legend()
plt.show()

# Ket : (Tipe Menara)
# 0 = 3 Kaki
# 1 = 4 Kaki
# 2 = Monopole
# 3 = Pole
```
